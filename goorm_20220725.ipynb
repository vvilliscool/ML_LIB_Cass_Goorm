## 머신러닝

```python
import sklearn
```

```python
import matplotlib.pyplot as plt
import scipy
import os
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.keys())
```

```python
print( iris.feature_names) # 입력이 어떤 데이터이다. feature 이름(컬럼 이름)
print( iris.data )  # 입력 - target를 예측하기 위해 사용하는 데이터()
```

```python

print( iris.target_names) # 입력이 어떤 데이터이다. feature 이름(컬럼 이름)
print( iris.target ) # 출력(target) - 예측해야 하는 feature(변수)
```

```python
print( iris.DESCR ) # 출력(target) - 예측해야 하는 feature(변수)
```

```python
from sklearn.model_selection import train_test_split

X = iris['data']
y = iris['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# 데이터 사이즈
print(X_train.shape)   # 훈련 데이터 셋 사이즈
print(X_test.shape)    # 테스트 데이터 셋 사이즈
print(y_train.shape)   # 훈련 데이터 레이블 사이즈 
print(y_test.shape)    # 테스트 데이터 레이블 사이즈
```

```python
import seaborn as sns
print(iris['feature_names'])  # 붓꽃의 꽃잎과 꽃받침의 feature 이름 
```

```python
iris_df = pd.DataFrame(X_train, columns=iris.feature_names)
iris_df['species'] = y_train
iris_df['species'] = iris_df['species'].astype('category') # 자료형 변환
iris_df.head()
```

```python
sns.pairplot(iris_df.iloc[ : ,0:4])   # 1~4열 선택
```

## First ML model

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model
```

```python
model.fit(X_train, y_train)
```

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
```

```python
### 예측시키기
pred = model.predict(X_new)
pred_targetname = iris['target_names'][pred]
print("예측 : ", pred)
print("예측한 타깃의 이름: ", pred_targetname)
```
