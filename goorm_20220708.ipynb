### 2-2 실습
  x의 범위는 0~100까지 짝수만, y값을 y*2로 만들고, 이를 선그래프로 표시.

```python
x = [ a for a in range(0,101,2)]
y = [ b * 2  for b in x  ]
print(y)

plt.plot(x, y, 'ro')
plt.show()
```

### 2-3 실습 2 
 (2) x의 범위는 0\~100까지 홀수만, y값을 x* x로 만들고, 이를 선그래프로 표시.

```python
x = [ a for a in range(1,101,2)]
y = [ b * b  for b in x  ]
print(y)

plt.plot(x, y, 'ro')
plt.show()
```

### 2-4 도전실습
  0~50까지 x의 범위, 
  y는 x의 값 중에 짝수만 리스트로 만들기
  이를 출력

```python
x = range(0, 50)
y = [v if v % 2 == 0 else v-1 for v in x  ]
plt.plot(x, y, 'ro')
plt.show()
```

```python
x = range(0, 15)
y = [v*v for v in x]
print(y)
fig, axs = plt.subplots(2, 2, figsize=(10,10))
axs[0, 0].plot(x, y, 'o')
axs[0, 1].plot(x, y, '-')
axs[1, 0].bar(x, y)
axs[1, 1].hist(x)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2 * np.pi, 0.1)
sin_y = np.sin(x)
cos_y = np.cos(x)
tan_y = np.tan(x)
arctan_y = np.arctan(x)

fig, axs = plt.subplots(2,2, figsize=(10,10))
axs[0,0].plot(x, sin_y, 'b--')
axs[0,0].set_title('Sin')
axs[0,0].set_ylabel('y')
axs[0,0].set_xlabel('x')

axs[0,1].plot(x, cos_y, 'r--')
axs[0,1].set_title('cos')
axs[0,1].set_ylabel('y')
axs[0,1].set_xlabel('x')

axs[1,0].plot(x, tan_y, 'b--')
axs[1,0].set_title('tan')
axs[1,0].set_ylabel('y')
axs[1,0].set_xlabel('x')

axs[1,1].plot(x, arctan_y, 'r--')
axs[1,1].set_title('arctan')
axs[1,1].set_ylabel('y')
axs[1,1].set_xlabel('x')
plt.show()
```

```python
for i in range(5):
    plt.plot(x, i*x, label='y = %ix' % i)

plt.legend() # 범례 추가
plt.show()
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2,3,1)
plt.plot(x, 1*x, label='y = 1x')
plt.legend()
plt.title("y=1x graph")

plt.subplot(2,3,2)
plt.plot(x, 2*x, label='y = 2x')
plt.legend()
plt.title("y=2x graph")

plt.subplot(2,3,3)
plt.plot(x, 3*x, label='y = 3x')
plt.legend()
plt.title("y=3x graph")

plt.subplot(2,3,4)
plt.plot(x, 4*x, label='y = 4x')
plt.legend()
plt.title("y=4x graph")

plt.subplot(2,3,5)
plt.plot(x, 5*x, label='y = 5x')
plt.legend()
plt.title("y=5x graph")
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2,2,1)
plt.plot(x, np.sin(x), label='y=sin(x)')
plt.legend()
plt.title("y=sin(x) graph")

plt.subplot(2,2,2)
plt.plot(x, np.cos(x), label='y=cos(x)')
plt.legend()
plt.title("y=cos(x) graph")

plt.subplot(2,2,3)
plt.plot(x, np.tan(x), label='y=tan(x)')
plt.legend()
plt.title("y=tan(x) graph")

plt.subplot(2,2,4)
plt.plot(x, np.arctan(x), label='y=arctan(x)')
plt.legend()
plt.title("y=arctan(x) graph")
```
