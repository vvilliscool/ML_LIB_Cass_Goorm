```python
team_score = { "toto":[1500,3000,5000,7000,5500],
               "apple":[4000,5000,6000,5500,4500],
               "gildong":[2000,2500,3000,4000,3000],
               "catanddog":[7000,5000,3000,5000,4000],
              "book": [5000,4000,3000,4000,5000]}

team_df = DataFrame(team_score)
team_df
```

```python
date = pd.date_range('2022-05-01', '2022-05-05')
team_df = pd.DataFrame(team_score, index=date)
team_df
```

```python
team_df.sum()  # 열의 합
```

```python
tmp_sum = team_df.sum(axis=1)
```

```python
team_df['Sum'] = tmp_sum
team_df
```

```python
tmp_mean = team_df[['toto','apple','gildong','catanddog']].mean(axis=1)
team_df['Mean'] = tmp_mean
team_df
```

```python
tmp_cnt = team_df[['toto','apple','gildong','catanddog']].count(axis=1)
team_df['Count'] = tmp_cnt
team_df
```
